```python
from PIL import Image

# Open the image file
img = Image.open("killua.jpg")

# Convert the image to grayscale
gray_img = img.convert('L')

width, height = gray_img.size

# Display the grayscale image with dimensions
plt.imshow(gray_img, cmap='gray')
plt.title(f"Grayscale Image ({width} x {height})")
plt.show()
```
